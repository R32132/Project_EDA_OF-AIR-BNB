```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have your DataFrame 'hotel_booking_df'

# Create a cross-tabulation table
room_type_by_neighbourhood = pd.crosstab(hotel_booking_df['neighbourhood_group'], hotel_booking_df['room_type'])

# Plot the stacked bar chart
room_type_by_neighbourhood.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Room Type Distribution by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Number of Bookings')
plt.legend(title='Room Type')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have your DataFrame 'hotel_booking_df'

# Create the scatter plot
sns.scatterplot(x='number_of_reviews', y='price', data=hotel_booking_df)
plt.title('Price vs. Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Price')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have your DataFrame 'hotel_booking_df'

# Create the box plot
sns.boxplot(x='room_type', y='price', data=hotel_booking_df)
plt.title('Price Distribution by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have your DataFrame 'hotel_booking_df'

# Create the scatter plot
sns.scatterplot(x='calculated_host_listings_count', y='availability_365', data=hotel_booking_df)
plt.title('Availability vs. Calculated Host Listings Count')
plt.xlabel('Calculated Host Listings Count')
plt.ylabel('Availability (365 days)')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have your DataFrame 'hotel_booking_df'

plt.figure(figsize=(10, 6))  # Adjust figure size if needed
sns.boxplot(x='neighbourhood_group', y='minimum_nights', data=hotel_booking_df)
plt.title('Minimum Nights Distribution by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Minimum Nights')
plt.show()
```
